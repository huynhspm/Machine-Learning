# **Image Compression Project**

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
# https://machinelearningcoban.com/assets/kmeans/girl3.jpg
img = mpimg.imread('Picture/girl3.jpg')
plt.imshow(img)
plt.axis('off')
plt.show() 
```

```python
print(img.shape)
# change array of image (height, width, rgb) to array(height * width, rgb)
X = img.reshape((img.shape[0]*img.shape[1], img.shape[2]))
print(X.shape)
```

```python
for K in [2, 5, 10, 15]:
    kmeans = KMeans(n_clusters = K).fit(X)
    labels = kmeans.predict(X)
    # np.zeros_like(X) is an other way
    imgArray = np.zeros(X.shape, dtype = int)
    for k in range(K):
        imgArray[labels == k] = kmeans.cluster_centers_[k]
    imgCompression = imgArray.reshape(img.shape[0], img.shape[1], img.shape[2])

    plt.imshow(imgCompression)
    plt.axis('off')
    plt.show()
```
